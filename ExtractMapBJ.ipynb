```python
# Read the csv file with all blocks
blocks = pd.read_csv('sig_cdmx_BENITO JUAREZ_08-2020.csv')

# Extract fids for Narvarte Poniente
benito_juarez = blocks[blocks["colonia_predio"] == "NARVARTE PONIENTE"]
benito_juarez.head(3)
```

```python
# Merge to extract polygons only for Narvarte Poniente
merged = polygons[polygons["fid"].isin(benito_juarez["fid"])]
merged.head(3)
```

```python
# Draw the polygons of Narvarte Poniente
ax3 = merged.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
```
